```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.metrics import mean_squared_error, r2_score

```

```python
df = pd.read_csv('mat2.csv')
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nTarget (G3) range:", df['G3'].min(), "to", df['G3'].max())
print("\nFirst 3 rows:")
df.head(3)
```

```python
df = df.drop('Unnamed: 0', axis=1, errors='ignore')
target = 'G3'

cat_cols = df.select_dtypes(include=['object']).columns.tolist()


label_encoders = {}
for col in cat_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le
print("✅ Encoding done!")
```

```python
X = df.drop(target, axis=1)
y = df[target]

selector = SelectKBest(score_func=f_regression, k=8)
X_selected = selector.fit_transform(X, y)
selected_features = X.columns[selector.get_support()].tolist()

print("🎯 TOP 8 FEATURES:")
for i, feat in enumerate(selected_features, 1):
    print(f"{i}. {feat}")
    
X_sel = pd.DataFrame(X_selected, columns=selected_features, index=X.index)
selected_features  # This displays the list
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_sel, y, test_size=0.2, random_state=42
)
print(f"Training: {X_train.shape}, Test: {X_test.shape}")
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

rf = RandomForestRegressor(random_state=42, n_jobs=-1)
param_dist = {
    'n_estimators': [100, 200], 
    'max_depth': [10, 20, None], 
    'min_samples_split': [2, 5]
}
search = RandomizedSearchCV(rf, param_dist, n_iter=10, cv=5, random_state=42, n_jobs=-1)
search.fit(X_train_scaled, y_train)

model = search.best_estimator_
print("✅ Model trained!")
print("Best params:", search.best_params_)
```

```python
y_pred = model.predict(X_test_scaled)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"📈 RMSE: {rmse:.2f}")
print(f"📈 R²:   {r2:.3f}")

# FIXED: Use ascending=False instead of descending=True
importances = model.feature_importances_
feat_df = pd.DataFrame({
    'feature': selected_features, 
    'importance': importances
}).sort_values('importance', ascending=False)  # ← FIXED!

print("\n🎯 Most important feature:")
print(feat_df.head(1)['feature'].iloc[0])
print("\nAll features:")
feat_df  # Displays table
```

```python
feat_df = feat_df.sort_values('importance', ascending=True)
plt.figure(figsize=(10, 6))
sns.barplot(data=feat_df, x='importance', y='feature')
plt.title('Top 8 Feature Importances')
plt.tight_layout()
plt.savefig('feature_importance.png', dpi=300)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.7, edgecolors='black')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('Actual G3'); plt.ylabel('Predicted G3')
plt.title('Actual vs Predicted Grades')
plt.savefig('actual_vs_predicted.png', dpi=300)
plt.show()
```

```python
def predict_grade(**kwargs):
    """Predict ANY student's grade! 🎯"""
    print("\n" + "="*50)
    print("🔮 NEW STUDENT PREDICTION")
    print("="*50)
    
    pred_data = {}
    for feature in selected_features:
        if feature in kwargs:
            pred_data[feature] = kwargs[feature]
            print(f"  {feature}: {kwargs[feature]} ✓")
        else:
            pred_data[feature] = X_sel[feature].median()
            print(f"  {feature}: [median: {pred_data[feature]:.1f}]")
    
    pred_df = pd.DataFrame([pred_data])[selected_features]
    pred_scaled = scaler.transform(pred_df)
    prediction = model.predict(pred_scaled)[0]
    
    print(f"\n PREDICTED G3: {prediction:.1f}/20")
    print("   (0-9=FAIL, 10-13=PASS, 14-20=GOOD)")
    return prediction

# 🧪 TEST IT:
predict_grade(G2=14, G1=12, failures=0)
```
